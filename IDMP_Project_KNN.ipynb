Data Loading and Preprocessing

```python
X_train = pd.read_csv('data_train.csv')
X_test = pd.read_csv('data_test.csv')
```

```python
Y_train = X_train['y']
Y_test = X_test['y']
```

```python
X_train = X_train.drop('y',axis=1)
X_train = X_train.drop('Unnamed: 0',axis=1)

X_test = X_test.drop('y',axis=1)
X_test = X_test.drop('Unnamed: 0',axis=1)
```

```python
X_test.head(1)
```

# With SMOTE

```python
Counter(Y_train)
```

```python
oversample = SMOTE()
X, y = oversample.fit_resample(X_train, Y_train)
```

```python
Counter(y)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=6, weights='uniform', algorithm='ball_tree')

neigh.fit(X, y)
```

```python
y_pred = neigh.predict(X_test)
print(f1_score(Y_test,y_pred))
print(precision_score(Y_test,y_pred))
print(recall_score(Y_test,y_pred))
print(accuracy_score(Y_test,y_pred))
```

```python
mean(cross_val_score(neigh, X, y, cv=10,scoring='f1'))
```

```python
from sklearn.neighbors import KNeighborsClassifier

Training= {}
Testing= {}
for weights in ['uniform', 'distance']:
  for algo in ['auto', 'ball_tree', 'kd_tree', 'brute']:
    for neighbours in range(1,10):
      neigh = KNeighborsClassifier(n_neighbors=neighbours, weights=weights, algorithm=algo)
      neigh.fit(X, y)

      scores = cross_val_score(neigh, X, y, cv=10,scoring='f1')
      s = str(weights)+" "+str(algo)+" "+str(neighbours)
      Training[s]= mean(scores)

      y_pred = neigh.predict(X_test)
      Testing[s]=f1_score(Y_test, y_pred)
      print(s)
```

```python
Training_List=[]
Training_List_names = []
for i in Training.items():
  Training_List_names.append(i[0])
  Training_List.append(i[1])

d_all_training = {"Model Config":Training_List_names,'Training Average CV F1':Training_List}  

Testing_List=[]
Testing_List_names = []
for i in Testing.items():
  Testing_List_names.append(i[0])
  Testing_List.append(i[1])

d_all_testing= {"Model Config":Testing_List_names,'Testing F1':Testing_List}
```

```python
df1= pd.DataFrame.from_dict(d_all_training)
df2 = pd.DataFrame.from_dict(d_all_testing)

```

```python
result = pd.concat([df1, df2], axis=1)
```
